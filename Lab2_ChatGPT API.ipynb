## 生成逐字稿(Transcription)

### 長語音：截短送至另一個檔案

```python
! pip3 install pydub
```

```python
from pydub import AudioSegment

song = AudioSegment.from_mp3("./record.wav")

# PyDub 以毫秒為單位
one_minutes = 1 * 60 * 1000
first_1_minutes = song[:one_minutes]
first_1_minutes.export("./record_1.wav", format="wav")
```

```python
two_minutes = 2 * 60 * 1000
second_minutes = song[one_minutes:two_minutes]
second_minutes.export("./record_2.wav", format="wav")
```

```python
audio_file= open("./record_1.wav", "rb")
transcript = openai.Audio.translate("whisper-1", audio_file)
transcript
```
